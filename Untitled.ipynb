```python
txtFile = open("./docs/D14.txt", "r")

coords = []
for lines in txtFile.readlines():
    coord_line = []
    for x in lines.split(" -> "):
        coord = []
        for y in x.split(','):
            coord.append(eval(y))
        coord_line.append(coord)
    coords.append(coord_line)

```

```python
smallest_x = 500
smallest_y = 100
biggest_x = 0
biggest_y = 0
for i, coord_line in enumerate(coords):
    for i, coord in enumerate(coord_line):
        if coord[0] > biggest_x:
            biggest_x = coord[0]
        
        if coord[0] < smallest_x:
            smallest_x = coord[0]
            
        if coord[1] > biggest_y:
            biggest_y = coord[1]
        
        if coord[1] < smallest_y:
            smallest_y = coord[1]   
        
print(smallest_x)
print(smallest_x-smallest_x, biggest_x-smallest_x)
print(smallest_y, biggest_y)
print(500 - smallest_x)
for line, coord_line in enumerate(coords):
    for i, coord in enumerate(coord_line):
        coords[line][i][0] -= smallest_x
```

```python
plane = [['.' for i in range(55)] for i in range(166)] 

def update_plane(start, end, plane):
    if start[0] == end[0]:
        if start[1] > end[1]:
            for i in range(start[1]-end[1]):
                plane[start[1]-i][start[0]] = '#'
        elif start[1] < end[1]:
            for i in range(end[1]-start[1]):
                plane[start[1]+i][start[0]] = '#'
    # line is horizontal            
    elif start[1] == end[1]:
        if start[0] > end[0]:
            for i in range(start[0]-end[0]):
                plane[start[1]][start[0]-i] = '#'
        elif start[0] < end[0]:
            for i in range(end[0]-start[0]):
                plane[start[1]][start[0]+i] = '#'
    return plane
        
for n_line, coord_line in enumerate(coords):
    for n_coord in range(len(coord_line)-1):
        update_plane(coord_line[n_coord], coord_line[n_coord+1], plane)
        #print(n_line, coord_line[n_coord], coord_line[n_coord+1])

units = 0
max_depth = 165

def try_left(pos):
    if plane[pos[1]+1][pos[0]-1] == '.':
        pos[1] += 1
        pos[0] -= 1
        return 1
    else:
        return 0
    
def try_right(pos):
    if plane[pos[1]+1][pos[0]+1] == '.':
        pos[1] += 1
        pos[0] += 1
        return 1
    else:
        return 0

def update_pos(curr_pos):
    while True:
        if curr_pos[1]+1 > max_depth:
            curr_pos[1]+=1
            return curr_pos
        if plane[curr_pos[1]+1][curr_pos[0]] == '.':
            curr_pos[1] += 1
        else:
            if not try_left(curr_pos):
                if not try_right(curr_pos):
                    return curr_pos
                    
        
    

    
while True:   
    unit_pos = [17,0]
    
    
    unit_pos = update_pos(unit_pos)
    if unit_pos[1] > max_depth:
        break
    elif unit_pos[0] < 0 or unit_pos[0] > 54:
        break
    else:
        plane[unit_pos[1]][unit_pos[0]] = 'o'
        units += 1

        

print(units)        
for i, line in enumerate(plane):
    print(','.join(line))
```

```python
def place_rocks(data):
    rocks = set()
    for line in data.split("\n"):
        points = [tuple(map(int, p.split(","))) for p in line.split(" -> ")]
        for i in range(len(points)-1):
            p1, p2 = points[i], points[i+1]
            xr = range(min(p1[0], p2[0]), max(p1[0], p2[0]) + 1)
            yr = range(min(p1[1], p2[1]), max(p1[1], p2[1]) + 1)
            rocks.update({(x, y) for x in xr for y in yr})
    return rocks

data = open("./docs/D14.txt").read().strip()
rocks = place_rocks(data)
max_y = max((y for _, y in rocks))
x, y = (500, 0)
ct = p1 = p2 = 0
while True:
    if (x, y) in rocks:  # restart sand at origin
        (x, y) = (500, 0)
    if y > max_y and p1 == 0:  # abyss part 1
        p1 = ct
    if (x, y + 1) not in rocks and y < max_y + 1:  # drop down?
        y += 1
    elif (x - 1, y + 1) not in rocks and y < max_y + 1:  # drop left-down?
        x -= 1
        y += 1
    elif (x + 1, y + 1) not in rocks and y < max_y + 1:  # drop right-down?
        x += 1
        y += 1
    else:  # hit somoething
        ct += 1
        rocks.add((x, y))
    if (x, y) == (500, 0):  # filled
        p2 = ct
        break
print(f"Part 1: {p1}")
print(f"Part 2: {p2}")
```
